## Introduction and data preparation

Our goal is to predict the probability of survival of passengers on the Titanic using information that was available _before_ the ship took off on its fatal maiden voyage. This information consists of the following (columns) per passenger (row):

- `PassengerId` is superfluous since we can track row index;
- `Survived` is a binary (0 = no, 1 = yes) indication of whether the passenger survived the accident;
- `Pclass` is an indication (1 to 3 with 1 being the most expensive) of travel class;
- `Name` could be useful after some engineering (e.g. by extracting titles) but we will not do that;
- `Sex` -- 'male' or 'female', represented as 0 or 1 respectively in created column `SexInt`;
- `Age` in full years (with missing values replaced by median value in training set) -- divided by 10 in created column `AgeSlash10` to bring coefficient up to a level that can more easily be compared with others (also, it would probably be useful to bin these into child, adult, old);
- `SibSp` denotes the number of siblings or spouses from this passenger that are aboard;
- `Parch` denotes the number of children or parents from this passenger that are aboard (the siblings/spouses and parents/children columns are poorly grouped, otherwise they could have been used to impute age data better -- said columns are now combined into `FamilyAboard`);
- `Ticket` is a hard-to-interpret code that seems superfluous considering the other information;
- `Fare` denotes the amount paid for the ticket (0-values are replaced with the (train set) median fare for the passenger's travel class) -- created column `LogFare` contains its natural logarithm that is less skewed and may prevent overfitting to one of the few very high values;
- `Cabin` is the often missing identifier of the cabin the passenger is staying in -- not very useful since the naming system is not obvious (which it would be if it denoted, e.g., part of ship and floor);
- `Embarked` denotes where the passenger got on -- either `C`herbourg, `Q`ueenstown, or `S`outhampton (replaced by -1, 0, and 1 to make everything numerical) (missing values are replaced by `S`/1, the most common port).

The data is split into a training part with 600 passengers and a test part with 291 passengers.

The classifier is assessed using 0/1-loss and sensitivity/specificity. A naive classifier that classifies all women as surviving and all men as dying serves as benchmark.

```python
%matplotlib inline
%precision 4

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt; plt.style.use('ggplot')
import seaborn as sns; sns.set()

from sklearn import metrics
from sklearn.model_selection import KFold
import pystan; pystan.__version__
```

```python
data = pd.read_csv('titanic.csv')

np.random.seed(2020) 
shuffled_indices = np.random.permutation(np.arange(data.shape[0]))
train_indices = shuffled_indices[:-291]
test_indices = shuffled_indices[-291:]
train_data = data.iloc[train_indices]
test_data = data.iloc[test_indices]

train_data.head(10)
```

```python
# These simple imputations and extra variables are briefly mentioned above.
data = data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
data['SexInt'] = data['Sex'].replace('male', 0).replace('female', 1)
data['FamilyAboard'] = (data['SibSp'] + data['Parch'] > 0).astype(int)
data['Embarked'] = data['Embarked'].replace('C', -1).replace('Q', 0)
data['Embarked'] = data['Embarked'].replace('S', 1).fillna(1) 

# The following use only training data.
data['AgeSlash10'] = data['Age'].fillna(train_data['Age'].median()) / 10
data['Tiny'] = (data['Age'] < 14).replace(False, 0).replace(True, 1)
median_fares = [train_data[train_data['Pclass'] == c]['Fare'].median() 
                for c in [1, 2, 3]]
for i, row in data[data['Fare'] == 0].iterrows():
    data.loc[i, 'Fare'] = median_fares[row['Pclass'] - 1]
data['LogFare'] = np.log(data['Fare'])
```

We can also do some rescaling based on observations of the train data.

This is only useful when considering all input variables as real instead of allowing all categories to have their own parameter.

```python
sns.barplot('Pclass', 'Survived', data=train_data); plt.show()
data['Pclass'] = data['Pclass'].replace(3, 0).replace(2, 4).replace(1, 6)
# Specific numbers are not necessarily perfect but at least not in [1, 2, 3]
# to avoid problems when running this cell multiple times.
sns.barplot('Embarked', 'Survived', data=train_data); plt.show()
data['Embarked'] = data['Embarked'].replace(-1, -4)
```

We can also add some interactions.

```python
data['SexClass'] = data['SexInt'] * data['Pclass']
data['SexFare'] = data['SexInt'] * data['LogFare']
data['SexAge'] = data['SexInt'] * data['AgeSlash10']
data['AgeClass'] = data['AgeSlash10'] * data['Pclass']
```

For a final

```python
train_data = data.iloc[train_indices]
test_data = data.iloc[test_indices]
y_train = train_data['Survived'].to_numpy()
y_test = test_data['Survived'].to_numpy()
train_data.head(10)
```

```python
def assess(y_pred, y_true=None):
    if y_true == None:
        if y_pred.shape[0] == 291:
            print("Evaluating on apparently also put-in test data!!!")
            y_true = test_data['Survived'].to_numpy()
        elif y_pred.shape[0] == 600:
            print("Evaluating on apparently also put-in train data!!!")
            y_true = train_data['Survived'].to_numpy()
        elif y_pred.shape[0] == 891:
            print("Evaluating on apparently also put-in _all_ data!!!")
            y_true = data['Survived'].to_numpy()

    y_pred_binary = np.array([1 if y > .5 else 0 for y in y_pred])
    accuracy = np.mean(y_pred_binary == y_true)

    tp = np.where(y_true == 1)[0]
    tn = np.where(y_true == 0)[0]

    sensitivity = sum(y_pred_binary[tp] == y_true[tp]) / len(y_true[tp])
    specificity = sum(y_pred_binary[tn] == y_true[tn]) / len(y_true[tn])

    # We only have one threshold, 0.5
    fpr_list, tpr_list, t = metrics.roc_curve(y_true, y_pred)
    auc = metrics.roc_auc_score(y_true, y_pred)
    plt.figure()
    plt.plot(fpr_list, tpr_list)
    plt.title(f'Area under curve = {auc}')
    plt.ylabel('True positive rate')
    plt.xlabel('False positive rate')
    plt.show()

    return {'0/1-loss': 1 - accuracy, 
            'sensitivity': sensitivity,
            'specificity': specificity,
            'auc': auc}


assess(train_data['SexInt'].to_numpy())
```

## Bayesian logistic regression

Let $x$ denote the information we have about a passenger _but_ $y$: whether s/he survived. $p(y|x)=\text{Bernoulli}(y|\mu(x))$ where $\mu(x) = \text{E}[y|x] = p(y = 1 | w, x) = 1 / (1 + e^{-w'x})$. $\sigma^2$ is chosen for prior $w \sim \text{Normal}(0, \sigma^2I)$ where $I$ (the square identity) has the same number of columns as $w$. That number is one higher than the number of columns originally in $x$, to accomodate an intercept. 

It makes some sense to use a flatter prior for the intercept, since its "true" value does not have to be close to zero. So $w_\texttt{Intercept} \sim \text{Normal}(0, (\sigma^2)^2)$.

We can actually be reasonably sure the posterior value is negative, since the mean survival rate is below .5. The story about the prior on, e.g., sex is similar -- there is a clear correlation between it and survival rate. But the explanations for these "facts" are not really a priori. If the Titanic had sunk under different circumstances, perhaps one woman would have died. 

Moving on with a sloppy (capitalizing) step to $n$ passengers, we want to find the $w$ that has maximal posterior $p(w | Y, X, \sigma^2) = p(Y|w, X) p(w) / p(Y | X, \sigma^2)$ where $p(Y | X, \sigma^2) = \int p(Y | w, X) p(w) dw$ is the hard integral that necessitates sampling. 

Then for any passenger $(x_\text{new},y_\text{new})$ (though if the passenger is really new/unseen we do not have $y_\text{new}$), we would predict $p(\hat{y}_\text{new} = 1| x_\text{new}, Y, X) = \text{E}_{p(w | Y, X, \sigma^2)} p(\hat{y}_\text{new} = 1 | x_\text{new}, w) = \int p(\hat{y}_\text{new} = 1 | x_\text{new}, w) p(w|Y, X) dw$, using the approximated distribution from before.

The interpretation of $w$ is a little tricky because $p(y = 1 | w, x) = 1 / (1 + e^{-w'x})$. The 0.5 decision boundary is "crossed" when $w'x = 0$, and while $p(y = 1 | w, x)$ increases along with $w'x$, the relationship is not linear. 

We use PyStan's No-U-Turn Sampling (NUTS) [1]. The proposal distribution is not static as in the Metropolis-Hasting algorithm. We leave most of Pystan's settings on default, set $\sigma^2$ as directly below (`sigma_squared`) somewhat arbitrarily, and experiment with combinations of variables and numbers of iterations. The post-warm-up chains are averaged when extraced from the model, but they are not permuted (as the package almost forces us to do). 

[1] M.D. Hoffman and A. Gelman. [The No-U-Turn Sampler](http://www.stat.columbia.edu/~gelman/research/published/nuts.pdf). Journal of Machine Learning Research 15 (2014) 1351-1381.

In this simple situation, relatively long runs are possible. On Google Colab, 10k extra iterations take about a minute. Traces look approximately the same over a 100k run with a lot of variables.

```python
sigma_squared = 10
```

```python
stan_code = """
data {
    int N;
    int M;
    matrix[N, M] X;
    int<lower=0, upper=1> y[N];
    int sigma_squared;
    int N2;
    int M2;
    matrix[N2, M2] X2;
}
parameters {
    real beta0;
    vector[M] beta; 
}
model {
    beta0 ~ normal(0, sigma_squared * sigma_squared);
    to_vector(beta) ~ normal(0, sigma_squared);
    y ~ bernoulli_logit(beta0 + X * beta);
}
generated quantities {
    vector[N2] y_pred;
    y_pred = beta0 + X2 * beta;
}
"""

stm = pystan.StanModel(model_code=stan_code)
```

### Variable selection

It's possible to test all combinations of variables that can be interpreted as real number. 

The last lecture by Bart seemed to go in a promising direction -- sampling a probability distribution for $k$, the number of variables to use -- but the did not seem very implementable to me. 

The following cross validation is not great in that regard -- overlapping aspects of the previously trained models are not taken into account. PyStan returns the posterior likelihood of the models, but it seems to correlate a lot with `auc`, area under the receiver operating curve (of the true versus false positive rate).

```python
def define_fit_assess(x_columns, iters_mult=200, n_folds=6):
    # This does rely on pre-processing (up to stm compilation) to have run!
    x_train = train_data[x_columns].to_numpy()
    print(x_columns)

    iters = 3500 + iters_mult * len(x_columns)
    # It might be an idea to shrink sigma_squared if there're more vars too.

    kf = KFold(n_splits=n_folds)
    results = []
    for train_index, test_index in kf.split(x_train):
        fx_train, fx_test = x_train[train_index], x_train[test_index]
        fy_train, fy_test = y_train[train_index], y_train[test_index]

        stan_data = {'N': fx_train.shape[0], 'M': fx_train.shape[1], 
                     'X': fx_train, 'y': fy_train, 
                     'sigma_squared': sigma_squared}

        stan_data = {'N': fx_train.shape[0], 'M': fx_train.shape[1], 
                     'X': fx_train, 'y': fy_train, 
                     'sigma_squared': sigma_squared,
                     'N2': fx_test.shape[0], 'M2': fx_test.shape[1], 
                     'X2': fx_test, 'y2': fy_test}

        fit = stm.sampling(data=stan_data, iter=iters, verbose=False)

        fy_pred = np.mean(fit.extract(permuted=False, 
                                      pars=['y_pred'])['y_pred'], axis=1)
        last_third_fy_pred_means = np.mean(fy_pred[-fy_pred.shape[0] // 3:], 
                                           axis=0)
        fy_test_hat = np.array([(1+np.exp(-x)) ** (-1) 
                                for x in last_third_fy_pred_means])
        fy_test_hat_b = np.array([1 if y > .5 else 0 
                                  for y in fy_test_hat])
        loss = 1 - metrics.accuracy_score(fy_test, fy_test_hat_b)
        auc = metrics.roc_auc_score(fy_test, fy_test_hat)
        results.append([loss, auc])
    print(np.mean(results, axis=0), np.std(results, axis=0))
    return {'mean loss, mean auc' : np.mean(results, axis=0),
            'loss sdev, auc  sdev': np.std(results, axis=0)}

all_interesting_combos = [  
    # Only a slightly random selection is left over here;
    # `SexInt`, `Pclass`, and `Tiny` seem most informative.
    ['SexInt', 'Pclass'],
    ['SexInt', 'LogFare'],
    ['SexInt', 'Pclass', 'Tiny'],
    # ['SexInt', 'Pclass', 'LogFare'],
    # ['SexInt', 'Pclass', 'Embarked'],
    # ['SexInt', 'Pclass', 'LogFare', 'Embarked'],
    ['SexInt', 'Pclass', 'SexClass', 'Embarked', 'Tiny'],
    # ['SexInt', 'Pclass', 'SexClass', 'Embarked', 'FamilyAboard'],
    # ['SexInt', 'Pclass', 'Embarked', 'AgeSlash10'],
    # ['SexInt', 'Pclass', 'Embarked', 'AgeSlash10', 'SexAge'],
    # ['SexInt', 'Pclass', 'Embarked', 'AgeSlash10', 'SexAge', 'AgeClass'],
    # ['SexInt', 'Pclass', 'SexClass', 'Embarked'],
    # ['SexInt', 'Pclass', 'SexClass', 'Embarked', 'AgeSlash10', 'SexAge'],
    # ['SexInt', 'Pclass', 'LogFare', 'Embarked', 'AgeSlash10', 
    #  'SexClass', 'SexAge', 'AgeClass', 'Tiny'],
]
results = [define_fit_assess(x_columns) for x_columns in all_interesting_combos]
```

```python
x_columns = ['SexInt', 'Pclass', 'SexClass', 'Embarked', 'Tiny']
# x_columns = ['SexInt', 'Pclass', 'LogFare', 'Embarked', 'AgeSlash10', 
#              'SexClass', 'SexAge', 'AgeClass', 'Tiny']
x_train = train_data[x_columns].to_numpy()
x_test = test_data[x_columns].to_numpy()

stan_data = {'N': x_train.shape[0], 'M': x_train.shape[1], 
             'X': x_train, 'y': y_train, 'sigma_squared': sigma_squared,
             'N2': x_test.shape[0], 'M2': x_test.shape[1], 
             'X2': x_test, 'y2': y_test}

fit = stm.sampling(data=stan_data, iter=int(10E3), verbose=True)
```

```python
# Trace plots from https://qiita.com/takeshikondo/items/1d45e02ab8d85faadf47
# Unfortunately, these _are_ permuted. 

la = fit.extract(pars=['beta0', 'beta'])
names = fit.model_pars
n_param = np.sum([1 if len(x) == 0 else x[0] for x in fit.par_dims])
# No y_pred:
names = names[:-1]
n_param -= y_test.shape[0]
f, axes = plt.subplots(n_param, 2, figsize=(10, 4*n_param))
cnt = 0
for name in names:
    dat = la[name]
    if dat.ndim == 2:
        for j in range(dat.shape[1]):
            d = dat[:,j]
            sns.distplot(d, hist=False, rug=True, ax=axes[cnt, 0])
            sns.lineplot(data=d,   alpha=0.8, lw=.5, ax=axes[cnt, 1])
            cnt += 1
    else:
        # Intercept
        sns.distplot(dat, hist=False, rug=True, ax=axes[cnt, 0])
        sns.lineplot(data=dat,   alpha=0.8, lw=.5, ax=axes[cnt, 1])
        cnt += 1

name_list = ['Intercept'] + x_columns

for i in range(2):
    for j, t in enumerate(name_list):
        axes[j, i].set_title(t)

[print(f"beta[{i + 1}] is for {c}") for i, c in enumerate(x_columns)]
print(fit.stansummary(pars=['beta0', 'beta']))
```
