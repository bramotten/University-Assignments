# Not really good results

## Task 1

```python
plainresultsLSTM = np.loadtxt(open("Results/Decimal/plainresultsLSTM.csv"), delimiter=",")
reverseresultsLSTM = np.loadtxt(open("Results/Decimal/reverseresultsLSTM.csv"), delimiter=",")
plainresultsGRU = np.loadtxt(open("Results/Decimal/plainresultsGRU.csv"), delimiter=",")
reverseresultsGRU = np.loadtxt(open("Results/Decimal/reverseresultsGRU.csv"), delimiter=",")
plainresultsSimpleRNN = np.loadtxt(open("Results/Decimal/plainresultsSimpleRNN.csv"), delimiter=",")
reverseresultsSimpleRNN = np.loadtxt(open("Results/Decimal/reverseresultsSimpleRNN.csv"), delimiter=",")
```

```python
def printRes(results, modeltype, order): 
    DIGITS=3
    TEST_SIZE = 10 ** (DIGITS * 2)
    corr_class = results[0]
    mse = results[1]
    mae = results[2]
    print(modeltype, order)
    print('Percentage of errors', modeltype, ' ', order, ': ', 100-(corr_class/TEST_SIZE*100), '\nMSE: ', mse, '\nMAE: ', mae)
```

```python
print('PLAIN')
printRes(plainresultsLSTM, 'LSTM', 'plain')
printRes(plainresultsGRU, 'GRU', 'plain')
printRes(plainresultsSimpleRNN, 'SimpleRNN', 'plain')

print('\nREVERSE')
printRes(reverseresultsLSTM, 'LSTM', 'reverse')
printRes(reverseresultsGRU, 'GRU', 'reverse')
printRes(reverseresultsSimpleRNN, 'SimpleRNN', 'reverse')
```
