We can center and scale the columns of $X$ so that (respectively) the columns have mean 0 and standard deviation 1. The result will be denoted by $\tilde{X}$ and $\tilde{x}_k$s.

````python
def center_and_scale(X, scaling=True):
    # Returns matrix with columns of X centered s.t. they have mean 0
    # and (if scaling) standard deviation 1.
    Xt = X.copy()
    Xt[:] -= Xt.mean(axis=0)
    if scaling:
        Xt[:] /= Xt.std(axis=0)
    return Xt


center_and_scale(X)[:3]
````

# Probabilistic PCA

## Idea

Sources for the following are mostly:
- the assignment description;
- the slides of 17 and 23 April;
- [Tipping, M. E. and C. M. Bishop (1999). Probabilistic principal component analysis. *Journal of the Royal Statistical Society*, Series B **21** (3), 611–622](http://www.robots.ox.ac.uk/~cvrg/hilary2006/ppca.pdf); and
- C.M. Bishop (1999). Variational principal components. *Proceedings Ninth International Conference on Artificial Neural Networks,* IEEE, **1**, 509–514.

The objective of PCA is to identify fewer than $p$ components (or axes or simply columns) that data can "effectively" be projected on. A maximal amount of variance should be retained if the original data is projected on these principal components -- a loss of variance can be considered a loss of information or a homogenization of the data. However, every added component increases the amount of retained variance. It is hard to decide what the ideal amount of components / retained variance is without a probabilistic quantification of the results.

Hence, the probabilistic PCA. It can give a probability assesment of the combination of data and components (via a latent variable model from which the data "follows"), so that:
- missing data can be dealt with;
- non-numerical data can be dealt with; and
- the likelihood of the data given the latent variable model is known. 

The goal is (still) to represent observed $p$-dimensional $x_1, \ldots, x_n$ by $d$-dimensional $z_1, \ldots, z_n$ with $d < p$. PPCA assumes $z_k \sim \mathcal{N}(0, I_d)$ and $x_k | z_k \sim \mathcal{N}(\mu + W z_k, \sigma^2I_p)$ or $\tilde{x}_k | z_k \sim \mathcal{N}(W z_k, \sigma^2I_p)$. We set $d$ and find the $p \times d$ "parameters" in $W$. 

(Note that (Tipping and) Bishop use(s) $t$ for observed and $x$ for latent -- here, these are $x$ and $z$ respectively as in the slides. $W$ is also tranposed and there must be other differences.)

We then have $\mathbb{E}(\tilde{x}_k) = 0$ (since $\mu = 0$ after centering) and $\operatorname{cov}(\tilde{x}_k) = WW' + \sigma^2I_p = C$. The marginal distribution $P(\tilde{x}_k) = \int P(\tilde{x}_k | z_k) P(z_k) \text{d}z_k$ is distributed $\mathcal{N}(0, C)$. The log likelihood of all of $\tilde{X}$ can even be simplified to $(- n / 2) [d \log(2\pi) + \log |C| + \text{trace}(C^{-1}S)]$. [Notes: the slides say $p$ instead of $d$; single bars denote determinant; and $S$ is defined below.] The maximum likelihood solutions don't have to be iteratively approached -- they are $$\hat{\sigma}^2 = 1 / (p - d) \sum^p_{k = d + 1} \lambda_k; \text{and}$$ $$\hat{W} = V_d(\Lambda_d - \hat{\sigma}^2 I_d)^{1/2}.$$ 

To obtain the variables used in the equations sample covariance $S = 1 / n \sum^n_{k=1} \tilde{x}_k' \tilde{x}_k$ has been decomposed into eigenvectors $V$ and corresponding eigenvalues $\Lambda$. $\Lambda$ is a diagonal matrix with values ($\lambda_k$s with $1 \leq k \leq p$) sorted from high to low. Subscripts on the above matrices indicate the number of columns of it that are used (starting at the first column).

(Note that we have set $d$ as we would have to in regular PCA, but with PPCA the likelihood of this choice _can_ be judged.)

With $\hat{M} = W'W + \sigma^2I_d$, the mean of the distribution of the latent conditional on observed $\mathbb{E}(z_k | \tilde{x}_k) = \tilde{x}_k \hat{W}' \hat{M}^{-1}$. Then $\hat{W} \mathbb{E}(z_k | \tilde{x}_k)$ is the reconstruction of $\tilde{x}_k$. The posterior mean is more skewed to 0 with a higher $\hat{\sigma}^2$. 

Tipping & Bishop (1999) note that the above is not an orthogonal projection of $\tilde{x}_k$ and is thus not expected to lead to the lowest possible squared error. The alternative projection (with $\mu=0$) is $\hat{W}(\hat{W}'\hat{W})^{-1}M \mathbb{E}(z_k | \tilde{x}_k)$.

## Implementation

````python
def ppca(X, d):
    # Performs PPCA with notation approximately as above.
    Xt = center_and_scale(X)
    n, p = Xt.shape
    if d > p:
        print("I will fail; d > p.")
        
    S = np.cov(Xt, rowvar=False, bias=True)
    lambdas, V = np.linalg.eig(S)
    
    sort_high_to_low = np.argsort(np.abs(lambdas))[::-1]
    lambdas = lambdas[sort_high_to_low]
    V = V[:, sort_high_to_low]  # NB: care about columns, not rows!
    
    sigmasq = 1 / max(p - d, 1) * sum(lambdas[d:])
    W = V[:, :d] @ np.sqrt(np.diag(lambdas[:d]) - sigmasq * np.identity(d))
    
    M = W.transpose() @ W + sigmasq * np.identity(d)
    Exp_Z = np.linalg.inv(M) @ W.transpose() @ Xt.transpose()
    reconstruct = np.transpose(W @ Exp_Z)
    
    mse_reconstruct = np.transpose(
        W @ np.linalg.inv(W.transpose() @ W) @ M  @ Exp_Z)

    C = W @ W.transpose() + sigmasq * np.identity(p)
    log_likelihood = - n / 2 * \
        (d * np.log(np.pi) + np.log(np.linalg.det(C)) + \
         np.trace(np.linalg.inv(C) @ S))
    
    return W, sigmasq, Exp_Z, reconstruct, log_likelihood, mse_reconstruct

ppca_on_X = ppca(X, 2)
print("W:")
print(ppca_on_X[0])
print("Sigma squared:", ppca_on_X[1])
print("Some reconstructions of X:")
print(ppca_on_X[3][:3])
print("Log likelihood:", ppca_on_X[4])
````

## Evaluation

````python
def mse(Xhat, X):
    return np.square(np.subtract(Xhat, X)).mean()


def evaluate(X, d, quiet=False):
    Xt = center_and_scale(X)
    W, _, _, ppca_reconstruct, ll, ppca_mse_reconstruct = ppca(Xt, d)    
    if not quiet:
        print(f"** Evaluating with d = {d} **")
        print("Probabilistic PCA's components (W):")
        print(W)
        print("Log likelihood:", ll)
        print("And mean squared error:", mse(ppca_reconstruct, Xt))
        print(f"Or, when reconstructing to maximize it:",
              mse(ppca_mse_reconstruct, Xt))

    pca = PCA(n_components=d)
    pca.fit(Xt)
    pca_reconstruct = pca.transform(Xt) @ pca.components_
    if not quiet:
        print("\nRegular PCA's components (W):")
        print(pca.components_.transpose())
        print("And mean squared error:", mse(pca_reconstruct, Xt))
    return ppca_reconstruct, ppca_mse_reconstruct, pca_reconstruct


evaluate(X, 2);
````

Evaluating on data we can know the true components of might be an idea. 

````python
def build_toy_dataset(N, D, K, sigma=1):
    # Source: http://edwardlib.org/tutorials/probabilistic-pca.
    x_train = np.zeros((D, N))
    w = np.random.normal(0.0, 2.0, size=(D, K))
    z = np.random.normal(0.0, 1.0, size=(K, N))
    mean = np.dot(w, z)
    for d in range(D):
        for n in range(N):
            x_train[d, n] = np.random.normal(mean[d, n], sigma)

    print(f"p={D},  true d={K}")
    if D < 8: print("True principal components:"); print(w)
    return center_and_scale(x_train.transpose())


P = 2; true_d = P - 1  # experiment with 
X2 = build_toy_dataset(9001, P, true_d)
print("Some of the data:")
print(X2[:3])
print()
X2ppca, X2ppca_mse, X2pca = evaluate(X2, true_d)
````

The likelihood may give us some insights about the best $d$.

````python
def get_d(X):
    llikelihoods = [round(ppca(X, d)[4], 1) for d in range(1, X.shape[1] + 1)]
    if len(llikelihoods) < 8: print("Log likelihoods:", llikelihoods)
    return np.argmax(llikelihoods) + 1  # llikelihoods[0] corresponds to d=1

print(f"Most likely d={get_d(build_toy_dataset(9001, 25, 4))}") 
print(f"---")
Xd = get_d(X)
print(f"For the car data we have d={Xd} (true is unknown)")
````

We see a lot more variance along the found "component(s)" in the (PPCA and PCA) reconstructions that are optimised for MSE. 

````python
def hinton(matrix, max_weight=None, ax=None):
    # Draws Hinton diagram for visualizing a weight matrix.
    # Source:
    # https://matplotlib.org/3.1.1/gallery/specialty_plots/hinton_demo.html.
    ax = ax if ax is not None else plt.gca()

    if not max_weight:
        max_weight = 2 ** np.ceil(np.log(np.abs(matrix).max()) / np.log(2))

    ax.patch.set_facecolor('gray')
    ax.set_aspect('equal', 'box')
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.yaxis.set_major_locator(plt.NullLocator())

    for (x, y), w in np.ndenumerate(matrix):
        color = 'white' if w > 0 else 'black'
        size = np.sqrt(np.abs(w) / max_weight)
        rect = plt.Rectangle([x - size / 2, y - size / 2], size, size,
                             facecolor=color, edgecolor=color)
        ax.add_patch(rect)

    ax.autoscale_view()
    ax.invert_yaxis()
    plt.title("Positive values are white, square size indicates magnitude.")

W2 = ppca(X2, true_d)[0]
print(W2)
hinton(W2)
````
